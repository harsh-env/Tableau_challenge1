```python
import glob
import os
```

```python
files_join = os.path.join("../raw_data", "JC-20*.csv")
files_list = glob.glob(files_join)
files_list
```

```python
df = pd.concat(map(pd.read_csv, files_list), ignore_index=False)
df.reset_index()
```

```python
count = df['member_casual'].value_counts()
count
```

```python
count = df['start_station_name'].value_counts()
count
```

```python
count = df['end_station_name'].value_counts()
count
```

```python

df['started_at']= pd.to_datetime(df['started_at'])
df['ended_at']= pd.to_datetime(df['ended_at'])
df.dtypes
```

```python
count = df['started_at'].value_counts()
count
```

```python
count = df['ended_at'].value_counts()
count
```

```python
df = df.rename(columns ={"rideable_type":"bike_type", "started_at":"starting_time","ended_at":"ending_time", 
                         "start_station_name": "Starting_Station", "end_station_name":"Ending_Station", "member_casual":"Member_Type"})
df.dtypes
```
